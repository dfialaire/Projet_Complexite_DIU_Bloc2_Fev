<h4><u>Comment créer des graphiques à partir de données, et les positionner à l'écran dans un notebook ? : </u></h4>

```python
from math import *
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def calculs_ordonnees(abcisses, modele):
    ordonnees=[]
    if modele=="lineaire": 
        ordonnees=abcisses
        ylabel_modele="y = x"
    elif modele=="quadratique":
        longueur_abcisses=len(abcisses)
        for i in range(longueur_abcisses):
            carre = abcisses[i]*abcisses[i]
            ordonnees.append(carre)
        ylabel_modele="y = x²"
    elif modele=="racine":
        longueur_abcisses=len(abcisses)
        for i in range(longueur_abcisses):
            racine= sqrt(abcisses[i])
            ordonnees.append(racine)
        ylabel_modele="y = x ^(1/2)"
    elif modele=="log":
        longueur_abcisses=len(abcisses)
        for i in range(longueur_abcisses):
            logi= log(abcisses[i])
            ordonnees.append(logi)
        ylabel_modele="y = log(x)"
    return ordonnees,ylabel_modele
def graphes(titre,abcisses,modele1,modele2,modele3,modele4):        
    fig = plt.figure(figsize=(12, 4)) # (largeur,hauteur) # en inches
    fig.suptitle(titre, fontsize=16)
    
    plt.subplot(1, 4, 1) # 1 ligne, 4 colonnes, position 1
    ordonnees,ylabel_modele=calculs_ordonnees(abcisses, modele1)
    plt.plot(abcisses, ordonnees, marker='+', color ='r') # r=red
    plt.ylabel(ylabel_modele)
    plt.xlabel('Titre de mes abcisses (unités)')

    plt.subplot(1, 4, 2) # 1 ligne, 4 colonnes, position 2
    ordonnees,ylabel_modele=calculs_ordonnees(abcisses, modele2)
    plt.plot(abcisses, ordonnees, marker='*', color ='b')  # b=blue
    plt.ylabel(ylabel_modele)
    plt.xlabel('Titre de mes abcisses (unités)')

    plt.subplot(1, 4, 3) # 1 ligne, 4 colonnes, position 3
    ordonnees,ylabel_modele=calculs_ordonnees(abcisses, modele3)
    plt.plot(abcisses, ordonnees, marker='x', color ='g') # g=green
    plt.ylabel(ylabel_modele)
    plt.xlabel('Titre de mes abcisses (unités)')
    
    plt.subplot(1, 4, 4) # 1 ligne, 4 colonnes, position 4
    ordonnees,ylabel_modele=calculs_ordonnees(abcisses, modele4)
    plt.plot(abcisses, ordonnees, marker='.', color ='m') # m=magenta
    plt.ylabel(ylabel_modele)
    plt.xlabel('Titre de mes abcisses (unités)')
    fig.subplots_adjust(wspace=1.1)
```

<u>Exemple : </u>

```python
abcisses=[4,9,16,25,36,49,64,81]
graphes("Ceci est le 1er graphique",abcisses,"lineaire","racine","quadratique","log")
graphes("Ceci est le 2nd graphique",abcisses,"quadratique","log","racine","lineaire")
```
