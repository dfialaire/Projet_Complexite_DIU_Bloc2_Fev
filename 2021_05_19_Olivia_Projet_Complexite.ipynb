<u><h2> 0 Le tri insertion : </h2></u>
<u><b>Principe : </b></u><br/>
L'algorithme principal du tri par insertion est un algorithme qui insère un élément dans une liste d'éléments déjà triés (par exemple, par ordre croissant).<br/>
Imaginez un joueur de cartes qui dispose de cartes numérotées. Il a des cartes triées (1 3 6 8) dans sa main gauche, et une carte 5 dans la main droite. Il faut placer la carte 5 après (1 3) et avant (6 8). <br/>
Pour trier entièrement un ensemble de cartes dans le désordre, il suffit alors de placer toutes ses cartes dans la main droite (la main gauche est donc vide), et d'insérer les cartes une à une dans la main gauche, en suivant la procédure ci-dessus.<br/><br/>
<u><b>Complexité : </b></u><br/>
Si n est la taille de la liste, le contenu de la première boucle est exécuté environ n fois (plus précisément, n-1 fois).<br/>
À chaque fois qu'on exécute cette première boucle, on exécute la deuxième boucle entre 0 et i fois : on part de la case i, et on va jusqu'à la case 0, en s'arrêtant avant si on trouve une carte plus petite que l'élément à insérer. On a donc un nombre de comparaisons variant entre 0 et i, donc entre 0 et n.<br/>
Le nombre total d'exécutions de la deuxième est donc plus grand que n * 0, et plus petit que n * n, entre 0 et n².<br/>
On dit donc que le nombre de calculs de cet algorithme est de l'ordre de n².<br/>
En pratique, cela signifie que si l'on double la taille du tableau, l'algorithme sera 4 fois plus lent, et si on la multiplie par 10, 100 fois plus lent.<br/>
En langage "scientifique", on dira que <b>la complexité du tri par insertion est de O(n²).</b><br/>
<br/>
<u><b>Complexité dans le pire cas : </b></u><br/>
Dans le pire des cas,le tableau est déjà classé à l'envers : il faut décaler à chaque fois toutes les cartes de la main gauche. À chaque tour de boucle on a donc i comparaisons (la taille de la main gauche).<br/>
Le nombre total de comparaisons est donc 1 + 2 + 3 + 4 + 5...+ n-1.<br/>
Combien vaut exactement ce nombre ? Appelons-le S. On a :

On a renversé l'addition sur la deuxième ligne, ce qui ne change pas la valeur de S.
En additionnant chaque somme des deux lignes "en colonnes", on additionne S + S, le résultat est 2S :

On voit que 2S, c'est n ajouté à lui-même n-1 fois (de 1 à n-1, il y a n-1 termes).<br/>
On a donc 2S = n(n-1) ou S = n(n-1)/2.<br/>
Quand n est très grand, n(n-1) est approximativement égal à n², et le nombre de comparaisons de l'algorithme est donc d'environ n²/2, ou n²*0.5 comparaisons.<br/>
Cependant, le facteur 0.5 n'a pas grand sens : sur un ordinateur deux fois plus rapide, on ira deux fois plus vite et sur un ordinateur 2 fois plus lent, deux fois moins vite. Pour estimer le temps mis par cet algorithme de manière indépendante de l'ordinateur, on dit donc que le nombre d'actions qu'il fait est "de l'ordre de n²". <b>La complexité  est donc en O(n²).</b>


<u><b>Complexité dans le meilleur des cas : </b></u><br/>
Dans le meilleur des cas, le tableau et déjà classé et l'algorithme effectura seulement n comparaisons. <b>La complexité  est donc en O(n).</b><br/>

```python
def tri_insertion(tableau):
    """ Tri par insertion
    entrée :
        tableau : liste à trier
    sortie :
        tableau : liste triée
        Nb_Action : entier qui compte le nombre d'actions effectuées"""
    Nb_Action=0
    n = len(tableau)
    Nb_Action = Nb_Action+1
    for i in range(1, n):
        Nb_Action = Nb_Action+1
        x = tableau[i] 
        Nb_Action = Nb_Action+1
        j = i
        Nb_Action = Nb_Action+1
        while j >0 and tableau[j-1] > x :
            Nb_Action = Nb_Action+1
            tableau[j] = tableau[j-1] 
            Nb_Action = Nb_Action+1
            j = j - 1 
            Nb_Action = Nb_Action+1
        tableau[j] = x 
        Nb_Action = Nb_Action+1
    return tableau, Nb_Action
    
liste_initiale=[12,4,7,20,3,9]
print("\nvoici la liste : ",liste_initiale)
liste_triee,Nb_Action_atom=tri_insertion(liste_initiale)

print("\nvoici la même liste triée: ",liste_triee)
print("\nNombre d'action atomique effectuées : ",Nb_Action_atom)
```

```python
from math import *
import random
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from timeit import default_timer as timer 
##################################################################################
def calculs_ordonnees_model(taille_tableaux,liste_Nb_Action, modele):
    """ 
    Fonction du cours qui renvoie les ordonnées du graphique en fonction du modèle linéaire ou quadratique
    """ 
    ordonnees=[]
    maxi_action=max(max(liste_Nb_Action),1)
    if modele=="lineaire": 
        longueur_abcisses=len(taille_tableaux)
        norme=maxi_action/(max(taille_tableaux))# NORMALISATION              
        for i in range(longueur_abcisses):
            lin=taille_tableaux[i]*norme   # NORMALISATION
            ordonnees.append(lin)
        ylabel_modele="y = x"
    elif modele=="quadratique":
        longueur_abcisses=len(taille_tableaux)
        norme=maxi_action/(max(taille_tableaux)**2)# NORMALISATION
        for i in range(longueur_abcisses):
            carre = taille_tableaux[i]**2*norme   # NORMALISATION
            ordonnees.append(carre)
        ylabel_modele="y = x²"
    elif modele=="nlogn":
        longueur_abcisses=len(taille_tableaux)
        norme=maxi_action/(max(taille_tableaux)*log(max(taille_tableaux)))# NORMALISATION
        for i in range(longueur_abcisses):
            nln = taille_tableaux[i]*log(taille_tableaux[i])*norme   # NORMALISATION
            ordonnees.append(nln)
        ylabel_modele="y = xlog(x)"        
    return ordonnees,ylabel_modele

##################################################################################
def Presentation_resultats(tri_souhaite,jeu_test,modele):
    """ 
        entrées :
            tri souhaite : insertion, bulle, sélection, rapide ou natif
            jeu_test : melangé, croissant ou décroissant
            modele : linéaire ou quadratique
        sortie :
            aucune
        
        Affiche les résultats dans trois fenêtres :  
            - A gauche : le calcul des temps d'exécution
            - Au milieu : le nombre d'actions effectuées
            - A droite : la courbe théorique
    """
    liste_temps_moy=[]
    liste_Nb_Action_moy=[]
    liste_taille_tab=[]
    Nb_essai=10
    ########################
    ## Création d'une suite de tableaux au hasard de tailles 2, 102, 202, ..., 1002
    ########################
    for taille_tab in range(2,1003,100):
        liste_Nb_Action=[]
        liste_temps=[]
        for essai in range(Nb_essai):
            tableau=random.sample(range(1,taille_tab+1),taille_tab)    #jeu_test=melangé         
            if jeu_test=="croissant":
                tableau.sort()
            elif jeu_test=="décroissant":
                tableau.sort(reverse=True)    
            ########################
            ## Réalisation des tris de chacun de ces tableaux et mesure des temps
            ########################        
            debut = timer()
            if tri_souhaite=="insertion":
                tableau,Nb_Action=tri_insertion(tableau)
            elif tri_souhaite=="bulle":
                tableau,Nb_Action=tri_bulle(tableau)
            elif tri_souhaite=="selection":
                tableau,Nb_Action=tri_selection(tableau)
            elif tri_souhaite=="rapide_iteratif":
                tableau,Nb_Action=tri_rapide_iteratif(tableau)
            elif tri_souhaite=="timsort":
                tableau,Nb_Action=timsort(tableau)
            elif tri_souhaite=="natif_python":
                tableau,Nb_Action=tableau.sort(),0    
            fin = timer()
            temps = fin - debut
            liste_temps.append(temps)
            liste_Nb_Action.append(Nb_Action) 
        ########################
        ## Moyenne
        ########################  
        moyenne_temps=(sum(liste_temps))/Nb_essai 
        moyenne_Nb_Action=(sum(liste_Nb_Action))/Nb_essai
        liste_temps_moy.append(moyenne_temps)
        liste_Nb_Action_moy.append(moyenne_Nb_Action)
        liste_taille_tab.append(taille_tab)         
 
        
    fig = plt.figure(figsize=(15, 5)) # (largeur,hauteur) # en inches
    titre='Tri '+tri_souhaite+' avec jeu de test '+jeu_test
    fig.suptitle(titre, fontsize=12)
    
    plt.subplot(1, 3, 1) # 1 ligne, 3 colonnes, position 1
    plt.plot(liste_taille_tab, liste_temps_moy, marker='+', color ='r') # r=red
    plt.ylabel('Temps d''exécution (en secondes)')
    plt.xlabel('Taille des tableaux')

    plt.subplot(1, 3, 2) # 1 ligne, 3 colonnes, position 2
    plt.plot(liste_taille_tab, liste_Nb_Action_moy, marker='*', color ='b')  # b=blue
    plt.ylabel('Nombre d''actions atomiques')
    plt.xlabel('Taille des tableaux')

    plt.subplot(1, 3, 3) # 1 ligne, 3 colonnes, position 3
    ordonnees,ylabel_modele=calculs_ordonnees_model(liste_taille_tab,liste_Nb_Action, modele)
    plt.plot(liste_taille_tab, ordonnees, marker='x', color ='g') # g=green
    plt.ylabel(ylabel_modele)
    plt.xlabel('x')
    
    fig.subplots_adjust(wspace=0.8)         
    return liste_taille_tab, liste_temps_moy, liste_Nb_Action_moy

tri_souhaite="insertion"
jeu_test="melangé"
modele="quadratique"
insertion_mel_taille,insertion_mel_temps,insertion_mel_action=Presentation_resultats(tri_souhaite,jeu_test,modele)   
jeu_test="décroissant"
modele="quadratique"
insertion_decroiss_taille,insertion_decroiss_temps,insertion_decroiss_action=Presentation_resultats(tri_souhaite,jeu_test,modele) 
jeu_test="croissant"
modele="lineaire"
insertion_croiss_taille,insertion_croiss_temps,insertion_croiss_action=Presentation_resultats(tri_souhaite,jeu_test,modele) 
```

<u><h2> I Le tri bulle : </h2></u>
<u><b>Principe : </b></u><br/>
Le principe du tri à bulles est de comparer deux valeurs adjacentes et d'inverser leur position si elles sont mal placées. Lorsque l'on souhaite trier l'ensemble par ordre croissant, si un premier nombre x est plus grand qu'un deuxième nombre y et que l'on souhaite trier l'ensemble par ordre croissant, alors x et y sont mal placés et il faut les inverser. On parcourt ainsi la liste jusqu'à ce qu'on ait réalisé n-1 passages (n représentant le nombre de valeurs à trier) ou jusqu'à ce qu'il n'y ait plus rien à inverser lors du dernier passage.<br/>
On s'aperçoit qu'au premier passage, on place le plus grand élément de la liste au bout du tableau, au bon emplacement. Pour le passage suivant, nous ne sommes donc plus obligés de faire une comparaison avec le dernière élément donc à chaque passage, le nombre de valeurs à comparer diminue de 1.
<br/>
<br/>
<u><b>Complexité : </b></u><br/>
Le nombre de comparaisons "if tableau[en_cours] > tableau[en_cours + 1]" est une valeur qui ne dépend que de la longueur n de la liste (n est le  nombre d'éléments du tableau), ce nombre est égal au nombre de fois que les itérations s'exécutent. Le comptage montre que la boucle "for en_cours in range(0, n - passage):" s'exécute n-1 fois au premier passage, n-2 fois au deuxième passage, jusqu'à n-n=0 au dernier passage.<br/>
La complexité en nombre de comparaisons est égale à la somme des n-1 premiers entiers S=1 + 2 + 3 + 4 + 5...+ n-1 vue dans la partie précédente.<br/>
S = n(n-1)/2.<br/>
<b>La complexité en nombre de comparaison est de de l'ordre de n², que l'on écrit O(n²).</b><br/>

<u><b>Complexité dans le pire cas : </b></u><br/>
 Le pire cas est celui où le tableau est trié dans l'ordre inverse et donc chaque cellule doit être échangée, dans cette éventualité il y adonc autant d'échanges que de tests.<br/>
<b>La complexité au pire en nombre d'échanges est de l'ordre de n², que l'on écrit O(n²).</b>
<br/>

<u><b>Complexité dans le meilleur des cas : </b></u><br/>
Dans le meilleur des cas, avec des données déjà triées, l'algorithme effectura seulement n - 1 comparaisons. <br/>
<b>La complexité dans le meilleur des cas est donc en O(n).</b>
<br/>

<u><b>Référence : </b></u><br/>
<b>Script : </b>  http://lwh.free.fr/pages/algo/tri/tri_bulle.html <br/>
<b>Principe : </b>  https://pub.phyks.me/sdz/sdz/le-tri-a-bulles.html

```python
def tri_bulle(tableau):
    """ Tri bulle
    entrée :
        tableau : liste à trier
    sortie :
        tableau : liste triée
        Nb_Action : entier qui compte le nombre d'actions effectuées"""
    permutation = True
    passage = 0
    Nb_Action = 0
    Nb_Action = Nb_Action+1
    n = len(tableau)
    while permutation == True:
        Nb_Action = Nb_Action+1
        permutation = False
        Nb_Action = Nb_Action+1
        passage = passage + 1
        Nb_Action = Nb_Action+1
        for en_cours in range(0, n - passage):
            Nb_Action = Nb_Action+1
            if tableau[en_cours] > tableau[en_cours + 1]:
                Nb_Action = Nb_Action+1
                permutation = True
                Nb_Action = Nb_Action+1
                #On échange les deux élements
                temp=tableau[en_cours]
                Nb_Action = Nb_Action+1
                tableau[en_cours]=tableau[en_cours + 1]
                Nb_Action = Nb_Action+1
                tableau[en_cours + 1]=temp
                #tableau[en_cours], tableau[en_cours + 1] = tableau[en_cours + 1],tableau[en_cours]
                Nb_Action = Nb_Action+1
    return tableau,Nb_Action
    
liste_initiale=[12,4,7,20,3,9]
print("\nvoici la liste : ",liste_initiale)
liste_triee,Nb_Action=tri_bulle(liste_initiale)

print("\nvoici la même liste triée: ",liste_triee)
print("\nNombre d'action atomique effectuées : ",Nb_Action)
```

<u><h2> II Le tri sélection : </h2></u>
<u><b>Principe : </b></u><br/>
Le principe du tri par sélection est simple : il suffit de rechercher le plus petit élément, le placer en première position du tableau, recommencer avec le second plus petit, le placer en seconde position et ainsi de suite jusqu'à avoir parcouru la totalité du tableau.
<br/>

<u><b>Complexité : </b></u><br/>
À la première itération, on effectue n-1 comparaisons. À la ième itération, on effectue donc n-i comparaisons (puisque à chaque itération on décrémente la taille du tableau).
Le nombre total de comparaisons pour trier un tableau de taille n est donc la somme :<br/>
S = n-1 + n-2 + n-3 +...+ n-(n-1) = 1 + 2 + 3 + 4 + 5...+ n-1 = n(n-1)/2.<br/>
<b>La complexité en nombre de comparaison est de de l'ordre de n², que l'on écrit O(n²).</b><br/>

<u><b>Complexité dans le pire cas : </b></u><br/>
Il n'existe pas ici de meilleur des cas ou de pire des cas : quelque soit l'allure des données à trier, l'algorithme prend le même temps.
<br/>

<u><b>Complexité dans le meilleur des cas : </b></u><br/>
<br/>

<u><b>Référence : </b></u><br/>
<b>Script : </b>  http://lwh.free.fr/pages/algo/tri/tri_selection.html <br/>
<b>Principe : </b>  https://pub.phyks.me/sdz/sdz/le-tri-par-selection.html

```python
def tri_selection(tableau):
    """ Tri par sélection
    entrée :
        tableau : liste à trier
    sortie :
        tableau : liste triée
        Nb_Action : entier qui compte le nombre d'actions effectuées"""
    Nb_Action = 0
    n = len(tableau)
    Nb_Action += 1
    for en_cours in range(0,n):    
        Nb_Action += 1
        plus_petit = en_cours
        Nb_Action += 1
        for j in range(en_cours+1,n) :
            Nb_Action += 1
            if tableau[j] < tableau[plus_petit] :
                Nb_Action += 1
                plus_petit = j
                Nb_Action += 1
        if plus_petit>en_cours :
            #On échange les deux élements
            Nb_Action += 1
            temp = tableau[en_cours]
            Nb_Action += 1
            tableau[en_cours] = tableau[plus_petit]
            Nb_Action += 1
            tableau[plus_petit] = temp
            Nb_Action += 1
    return tableau,Nb_Action
    
liste_initiale=[12,4,7,20,3,9]
print("\nvoici la liste : ",liste_initiale)
liste_triee,Nb_Action=tri_selection(liste_initiale)

print("\nvoici la même liste triée: ",liste_triee)
print("\nNombre d'action atomique effectuées : ",Nb_Action)
```

<u><h2> III Le tri rapide non récursif : </h2></u>
<u><b>Principe : </b></u><br/>
Le tri rapide, ou quicksort, fait partie des tris qui utilisent les comparaisons. Son principe suit l'approche "diviser pour régner" : on sélectionne un élément de l'ensemble (le pivot) avant de placer d'un coté les éléments inférieurs ou égaux à cet élément, et de l'autre ceux qui lui sont supérieurs. On parle de partitionnement. On applique à nouveau le même procédé sur les deux sous-ensembles, et ainsi de suite. Il peut se résumer de la manière suivante : On partitionne l'ensemble en deux et on trie séparément les deux partitions.
<br/>

<u><b>Complexité : </b></u><br/>
Seule la littérature sur le tri rapide récursif existe.

```python
def partition(tableau,deb,fin,Nb_Action):
    """ Partitionne le tableau en deux sous tableaux :
    - celui de gauche contient les valeurs qui sont inférieures ou égales au pivot, 
    - celui de droite contient les valeurs supérieures au pivot.
    entrée :
        tableau : liste à trier
        deb : index de début
        fin : index de fin (notre pivot ici)
        Nb_Action : nombre d'actions effectuées
    sortie : 
        i+1 : index du pivot """      
    i = ( deb - 1 )
    Nb_Action += 1
    x = tableau[fin]
    Nb_Action += 1
  
    for j in range(deb , fin):
        Nb_Action += 1
        if   tableau[j] <= x:
            Nb_Action += 1
            # increment index of smaller element
            i = i+1
            Nb_Action += 1
            tableau[i],tableau[j] = tableau[j],tableau[i]
            Nb_Action += 1
  
    tableau[i+1],tableau[fin] = tableau[fin],tableau[i+1]
    Nb_Action += 1
    return (i+1,Nb_Action)
  
def tri_rapide_iteratif(tableau):
    """ Tri rapide itératif
    entrée :
        tableau : liste à trier
    sortie : 
        tableau : liste triée """ 
    Nb_Action=0
    # Crée la pile
    deb=0
    Nb_Action += 1
    fin=len(tableau)-1
    Nb_Action += 1
    taille = fin - deb + 1
    Nb_Action += 1
    pile = [0] * (taille)
    Nb_Action += 1
  
    # Initialise le pointeur de la pile
    pointeur = -1
    Nb_Action += 1
  
    # Empile les valeurs initiales de deb et fin dans la pile
    pointeur = pointeur + 1
    Nb_Action += 1
    pile[pointeur] = deb
    Nb_Action += 1
    pointeur = pointeur + 1
    Nb_Action += 1
    pile[pointeur] = fin
    Nb_Action += 1
  
    # Continue de dépiler la pile tant qu'elle est non vide
    while pointeur >= 0:
        Nb_Action += 1
  
        # Empile fin et deb
        fin = pile[pointeur]
        Nb_Action += 1
        pointeur = pointeur - 1
        Nb_Action += 1
        deb = pile[pointeur]
        Nb_Action += 1
        pointeur = pointeur - 1
        Nb_Action += 1
        
        # positionne le pivot à la bonne place dans le tableau trié
        pivot, Nb_Action = partition( tableau, deb, fin, Nb_Action)
            
        # S'il y a des éléments à gauche du pivot, on empile les bornes du tableau de gauche dans la pile
        if pivot-1 > deb:
            Nb_Action += 1
            pointeur = pointeur + 1
            Nb_Action += 1
            pile[pointeur] = deb
            Nb_Action += 1
            pointeur = pointeur + 1
            Nb_Action += 1
            pile[pointeur] = pivot - 1
            Nb_Action += 1
  
        # S'il y a des éléments à droite du pivot, on empile les bornes du tableau de droite dans la pile
        if pivot+1 < fin:
            Nb_Action += 1
            pointeur = pointeur + 1
            Nb_Action += 1
            pile[pointeur] = pivot + 1
            Nb_Action += 1
            pointeur = pointeur + 1
            Nb_Action += 1
            pile[pointeur] = fin
            Nb_Action += 1
    return tableau, Nb_Action
  
liste_initiale=[12,4,7,20,3,9]
print("\nvoici la liste : ",liste_initiale)
liste_triee,Nb_Action=tri_rapide_iteratif(liste_initiale)

print("\nvoici la même liste triée: ",liste_triee)
print("\nNombre d'action atomique effectuées : ",Nb_Action)

```

<u><h2> IV Le tri natif de Python : Timsort</h2></u>
<u><b>Principe : </b></u><br/>
L’algorithme de tri utilisé en Python est appelé timsort, inventé par Tim Peters en 2002.
L’idée principale est que dans un ensemble de données, une partie est déjà triée et qu’il faut en tirer un avantage. 
L'algorithme procède en parcourant le tableau à la recherche de monotonies, c'est-à-dire dans ce cas de suites d'au moins deux éléments consécutifs croissantes ou strictement décroissantes − les monotonies décroissantes sont simplement renversées, le caractère strict est donc indispensable pour garantir la stabilité de l'algorithme.
Pour des raisons de performance, une taille minimum M est fixée pour la longueur des monotonies recherchées : si la monotonie en cours de construction a en pratique une taille inférieure à M, c'est-à-dire si la suite des M éléments présents à cet endroit du tableau n'est pas monotone, une monotonie est artificiellement construite en effectuant un tri par insertion stable sur les Méléments consécutifs présents à l'endroit en question du tableau.
<br/>

<u><b>Complexité : </b></u><br/>
Dans le pire des cas, quand la liste est triée dans le sens inverse, Timsort a une complexité en O(n log(n)). <br/>
Dans le meilleur cas, quand la liste est déjà triée, la complexité est linéaire.
<br/>

<u><b>Référence : </b></u><br/>
<b>Script : </b>  https://www.geeksforgeeks.org/timsort/ <br/>
<b>Principe : </b> https://fr.wikipedia.org/wiki/Timsort

```python
# Python3 program to perform basic timSort
MIN_MERGE = 32
 
def calcMinRun(n,Nb_Action):
    """Returns the minimum length of a
    run from 23 - 64 so that
    the len(array)/minrun is less than or
    equal to a power of 2.
 
    e.g. 1=>1, ..., 63=>63, 64=>32, 65=>33,
    ..., 127=>64, 128=>32, ...
    """
    r = 0
    Nb_Action += 1
    while n >= MIN_MERGE:
        Nb_Action += 1
        r |= n & 1
        Nb_Action += 1
        n >>= 1
        Nb_Action += 1
    return n + r, Nb_Action
 


# This function sorts array from left index to
# to right index which is of size atmost RUN
def insertionSort(arr, left, right,Nb_Action):  
    for i in range(left + 1, right + 1):
        Nb_Action += 1
        j = i
        Nb_Action += 1
        while j > left and arr[j] < arr[j - 1]:
            Nb_Action += 1
            arr[j], arr[j - 1] = arr[j - 1], arr[j]
            Nb_Action += 2
            j -= 1
            Nb_Action += 1
    return Nb_Action
 

# Merge function merges the sorted runs
def merge(arr, l, m, r, Nb_Action):
     
    # original array is broken in two parts
    # left and right array
    len1, len2 = m - l + 1, r - m
    Nb_Action += 2
    left, right = [], []
    Nb_Action += 2
    for i in range(0, len1):
        Nb_Action += 1
        left.append(arr[l + i])
        Nb_Action += 1
    for i in range(0, len2):
        Nb_Action += 1
        right.append(arr[m + 1 + i])
        Nb_Action += 1
    i, j, k = 0, 0, l
    Nb_Action += 3
     
    # after comparing, we merge those two array
    # in larger sub array
    while i < len1 and j < len2:
        Nb_Action += 1
        if left[i] <= right[j]:
            Nb_Action += 1
            arr[k] = left[i]
            Nb_Action += 1
            i += 1
            Nb_Action += 1
        else:
            Nb_Action += 1
            arr[k] = right[j]
            Nb_Action += 1
            j += 1
            Nb_Action += 1
        k += 1
        Nb_Action += 1
 
    # Copy remaining elements of left, if any
    while i < len1:
        Nb_Action += 1
        arr[k] = left[i]
        Nb_Action += 1
        k += 1
        Nb_Action += 1
        i += 1
        Nb_Action += 1
 
    # Copy remaining element of right, if any
    while j < len2:
        Nb_Action += 1
        arr[k] = right[j]
        Nb_Action += 1
        k += 1
        Nb_Action += 1
        j += 1
        Nb_Action += 1
    return Nb_Action

        
# Iterative Timsort function to sort the
# array[0...n-1] (similar to merge sort)
def timsort(arr):
    Nb_Action=0
    n = len(arr)
    Nb_Action += 1
    minRun, Nb_Action = calcMinRun(n,Nb_Action)
     
    # Sort individual subarrays of size RUN
    for start in range(0, n, minRun):
        Nb_Action += 1
        end = min(start + minRun - 1, n - 1)
        Nb_Action += 1
        Nb_Action=insertionSort(arr, start, end,Nb_Action)

    # Start merging from size RUN (or 32). It will merge
    # to form size 64, then 128, 256 and so on ....
    size = minRun
    Nb_Action += 1
    while size < n:
        Nb_Action += 1 
        # Pick starting point of left sub array. We
        # are going to merge arr[left..left+size-1]
        # and arr[left+size, left+2*size-1]
        # After every merge, we increase left by 2*size
        for left in range(0, n, 2 * size):
            Nb_Action += 1
            # Find ending point of left sub array
            # mid+1 is starting point of right sub array
            mid = min(n - 1, left + size - 1)
            Nb_Action += 1
            right = min((left + 2 * size - 1), (n - 1))
            Nb_Action += 1
            # Merge sub array arr[left.....mid] &
            # arr[mid+1....right]
            if mid < right:
                Nb_Action += 1
                Nb_Action=merge(arr, left, mid, right,Nb_Action)
        size = 2 * size
        Nb_Action += 1
    return arr,Nb_Action

liste_initiale=[12,4,7,20,3,9]
print("\nvoici la liste : ",liste_initiale)
liste_triee,Nb_Action=timsort(liste_initiale)

print("\nvoici la même liste triée: ",liste_initiale)
print("\nNombre d'action atomique effectuées : ",Nb_Action)
```
